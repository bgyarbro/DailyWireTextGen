```python
def getArticleLinks(page = 1):
    """this function..."""
    data = requests.get('https://www.dailywire.com/authors/ben-shapiro?page=' + str(page))
    soup = BeautifulSoup(data.text, 'html.parser')
    article_links = []
    for article in soup.find_all('article'):
        for a in article.find_all('a'):
            article_links.append(a)
    article_links = article_links[0:-27]
    article_urls = [re.findall("/news/.*?>", str(article_link)) for article_link in article_links]
    article_urls_dropna = list(filter(None, article_urls)) # fastest
    article_urls_flat = [item for sublist in article_urls_dropna for item in sublist]
    article_urls_unique = list(set(article_urls_flat))
    article_urls_ends_stripped = [article[0:-2] for article in article_urls_unique]
    article_urls_clean = ['https://www.dailywire.com' + article for article in article_urls_ends_stripped]
    return(article_urls_clean)
```

```python
def scrapeArticle(article_link):
    data = requests.get(article_link)
    soup = BeautifulSoup(data.text, 'html.parser')
    
    articletext = ""
    for body in soup.find_all('div', { 'class': 'field-body' }):
        for p in body.find_all('p'):
            articletext = articletext + " " + p.text

    data = []
    div = soup.find('div', { 'class': 'field-body' })
    for p in div.find_all('p'):
        values = p.text
        data.append(values)

    return(data)
```
